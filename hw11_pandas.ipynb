### Задание 1

```python
import pandas as pd
```

```python
def class_(row):
    result = ''
    if row <= 2:
        result = 'Низкий рейтинг'
    elif row <= 4:
        result = 'Средний рейтинг'
    elif row == 4.5 or row == 5:
        result = 'Высокий рейтинг'
    return result
```

```python
ratings = pd.read_csv('ratings.csv')
ratings_by_film = ratings[['movieId', 'rating']].groupby('movieId').mean()
ratings_by_film['class'] = ratings_by_film['rating'].apply(class_)
ratings_by_film
```

### Задание 2

```python
geo_class = pd.read_csv('keywords.csv')
geo_class.head()
```

```python
def geo_class_f(row):
    geo_data = {'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}
    row = row.strip().split(' ')
    for el in row:
        if el in sum(geo_data.values(),[]):
            for key in geo_data:
                if el in geo_data[key]:
                    return key
    return 'Undefined'
        
```

```python
geo_class['region'] = geo_class['keyword'].apply(geo_class_f)
```

```python
geo_class.query('region != "Undefined"').head()
```

### Задание 3

```python
movies_df = pd.read_csv('movies.csv')
ratings_df = pd.read_csv('ratings.csv')
ratings_df.head()
```

```python
def production_year(row):
    years = list(range(1950,2011))
    for year in years:
        if str(year) in row:
            return year
    return 1900
```

```python
movies_df['release_year'] = movies_df['title'].apply(production_year)
movies_df.tail()
```

```python
result = (ratings_df.merge(movies_df, on='movieId', how = 'left').query('release_year != 1900')[['rating','release_year']]
.groupby('release_year').mean('rating').sort_values('rating', ascending=False))
```
