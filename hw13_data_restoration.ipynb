```python
#удаление выбросов
ol_indexes = list(pd.concat([data[np.abs(data.resp_r-data.resp_r.mean()) > (3*data.resp_r.std())], data[np.abs(data.pulse-data.pulse.mean()) > (3*data.pulse.std())]]).index)
data_without_ol = data.drop(ol_indexes)
data_without_ol
```

```python
data_without_ol[data_without_ol['surgery'].isna()]
```

```python
data_without_ol = data_without_ol.drop(labels = 131) #удаляем эту строку, так как в ней отсутствуют значения в ключевых столбцах
```

```python
data_without_ol['age'] = data_without_ol['age'].replace(9,2)
#по-видимому, в данных по возрасту категория 2 всегда отмечалась цифрой 9
```

### Задание 3

```python
data_without_ol.corr()
```

```python
data_without_ol.info() 
'''
Восстанавливать значения в столбцах rect_t, resp_r, temp_ext и pain будем по возрасту + пульсу, так как столбец возраст 
не имеет пропусков, а столбец pulse имеет наибольшее количество non-NaN значений, к тому же он состоит не из категориальных
значений. 
Помимо этого, в пользу данного решения говорит то, что коррелляция между столбцом pulse и восстанавливаемыми
столбцами максимальна в данном датафрейме.
'''
```

```python
data_without_ol['pulse'].fillna(data_without_ol.groupby(['age', 'rect_t'])['pulse'].transform('median'), inplace=True)
data_without_ol['pulse'].fillna(data_without_ol.groupby(['age', 'resp_r'])['pulse'].transform('median'), inplace=True)


data_without_ol['rect_t'].fillna(data_without_ol.groupby(['age', 'pulse'])['rect_t'].transform('median'), inplace=True)
data_without_ol['resp_r'].fillna(data_without_ol.groupby(['age', 'pulse'])['resp_r'].transform('median'), inplace=True)
data_without_ol['temp_ext'].fillna(data_without_ol.groupby(['age', 'pulse'])['temp_ext'].transform('median'), inplace=True)
data_without_ol['pain'].fillna(data_without_ol.groupby(['age', 'pulse'])['pain'].transform('median'), inplace=True)

#заполняем 
data_without_ol.info()
```

```python
data = data_without_ol.dropna(thresh = 8)
#остальные значения имеет смысл удалить, так как восстанавливать значения в pulse (и в других столбцах) по категориальным значениям не имеет смысла
#кроме того, все возможные значения в pulse, которые можно было восстановить не по категориальным данным уже были восстановлены
```
