```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
btc_data = btc_data.fillna(method = 'bfill') 
'''

Так как данные финансовые и едва ли возможны резкие скачки значений в течение одного-двух дней, 
а пропуски незначительны, то используем бэкфил. 
Лучше было бы заполнить значения средним арифметическим между предыдущим и последующим
не пустым значениями, не смог придумать адекватный способ это сделать.

''' 
```

```python
doge_data = pd.read_csv('DOGE-USD-2.csv')
eth_data = pd.read_csv('ETH-USD-2.csv')
display(doge_data[doge_data.isna().any(axis=1)])
display(eth_data[eth_data.isna().any(axis=1)])
```

```python
doge_data = doge_data.fillna(method = 'bfill')
eth_data = eth_data.fillna(method = 'bfill')
```

#### Так как в исходных данных достаточно мало столбцов для анализа, расчитаем ещё несколько метрик:

```python
def find_diff (row):
    '''
    Считает разницу в курсе за день
    '''
    return row['Close'] - row['Open'] 
    
    
def find_diff_perc (row):
    '''
    Считает разницу в курсе за день в процентах
    '''
    return round(row['Close']/row['Open']-1, 4)
```

```python
btc_data['Difference'] = btc_data.apply(find_diff, axis=1)
eth_data['Difference'] = eth_data.apply(find_diff, axis=1)
doge_data['Difference'] = doge_data.apply(find_diff, axis=1)

btc_data['Difference_perc'] = btc_data.apply(find_diff_perc, axis=1)
eth_data['Difference_perc'] = eth_data.apply(find_diff_perc, axis=1)
doge_data['Difference_perc'] = doge_data.apply(find_diff_perc, axis=1)
```

```python
btc_data['Gap'] = round(btc_data.Open.sub(btc_data.Close.shift(), fill_value=0), 2)
btc_data['Gap'][0] = 0

eth_data['Gap'] = round(eth_data.Open.sub(eth_data.Close.shift(), fill_value=0), 2)
eth_data['Gap'][0] = 0

doge_data['Gap'] = round(doge_data.Open.sub(doge_data.Close.shift(), fill_value=0), 2)
doge_data['Gap'][0] = 0

#Гэп - разница между курсом на момент закрытия и курсом на момент открытия следующего дня
```

```python
plt.plot(doge_data.Date, doge_data.Difference_perc)
plt.plot(eth_data.Date, eth_data.Difference_perc)
plt.plot(btc_data.Date, btc_data.Difference_perc)
plt.ylabel('Изменение курса за день', labelpad = 20)
plt.legend(['DOGE','ETH','BTC'])

plt.axhline(0, lw=1, color='black')

my_ticks = range(0, len(btc_data))[::30]
plt.xticks(ticks = my_ticks, rotation=45)
```

Из данного графика можно сделать вывод о том, что курсы трёх выбранных криптовалют меняются практически синхронно, 
однако, в процентных показателях курс догкоина зачастую реагирует на изменения на рынке гораздо более интенсивно.

```python
sns.pairplot(btc_data[['Open','Close','Volume', 'Difference', 'Gap', 'Difference_perc']])
```

- Прослеживается слабая корелляция между объёмом и ценой (как открытия, так и закрытия)
- Есть несколько интересных выбросов в столбце "Объём" и в столбце "Гэп"

```python
volume_data = btc_data[['Date', 'Volume']].merge(eth_data[['Date', 'Volume']],how='left', on='Date',suffixes =('_btc','_eth')).merge(doge_data[['Date', 'Volume']],how='left', on='Date')
volume_data = volume_data.rename(columns={volume_data.columns[3]: 'Volume_doge'})
volume_data.head(3) #Создадим датафрейм с объёмами торговли криптовалютами
```

```python
volume_data.plot(kind='area',
                             x = 'Date',
                             stacked=False,
                             title='Объёмы торговли криптовалютами', 
                             xlabel='Дата', 
                             ylabel='Объём, USD')
```

Пользуясь данной визуализацией, можно сделать вывод о том, что к биткоину относятся с большим доверием, так как его
объёмы торговли всегда выше, чем у других представленных на графике криптовалют, однако эта разница в объёмах имела тенденцию к снижению в апреле-июне 2021 года
